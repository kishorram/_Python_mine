```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv(r"https://raw.githubusercontent.com/Premalatha-success/Supervised-Learning/main/titanic-training-data.csv")
df
```

```python
df["Sex"].value_counts()
```

```python
sns.countplot(data=df,x='Sex')
```

```python
sns.countplot(data=df,x='Sex',hue='Survived')
```

```python
sns.countplot(data=df,x='Pclass',hue='Survived')
```

```python
sns.countplot(data=df,x='Embarked',hue='Survived')
```

```python
sns.relplot(x="Survived", y="Pclass", data = df)
```

```python
sns.relplot(x="Survived", y="Age", data = df)
```

```python
sns.relplot(x="Survived", y="Pclass", hue = "Sex", data = df);
```

```python
sns.relplot(x="Survived", y="Age", data = df);
```

```python
sns.relplot(x="Survived", y="Pclass", hue = "SibSp", data = df);
```

```python
sns.catplot(x="Sex", y="Age", data=df)
```

```python
sns.catplot(x="Sex", y="Age", jitter = False,  data=df)
```

```python
sns.catplot(x="Embarked", y="Age", hue = "Sex", data=df)
```

```python
sns.catplot(x="Sex", y="Age", kind = "swarm", data=df)
```

```python
sns.catplot(x="Sex", y="Age", hue = "SibSp", kind = "swarm", data=df)
```

```python
sns.catplot(x="Sex", y="Age", kind = "box", data=df)
```

```python
sns.catplot(x="Sex", y="Age", hue = "Embarked", kind = "box", data=df)
```
